Finding the disparity between how many popular apps are on the app store and how many unpopular apps are on the app store
3/4/2024
The dataset is a set of data about all kinds of app store statistics.


```python
import pandas as pd # Pandas works with python to help us wrangle data. It can sort, clean, and read our csv files. csv files are how we input data
import numpy as np # Numpy handles the math!
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Chazzyboi011/APCSP/main/googleplaystorecsv.csv")
data.tail(50)
```

```python
data.dropna(inplace = True)
```

```python
data.Price.hist()
```

```python
plt.hist(data.Price,bins=10)
```

```python
plt.hist(data.Rating,bins=10, range=(0,5))
```

```python
plt.hist(data.Installs,bins=20, rwidth=5)
#This Histogram shows that there are primarily popular apps
```

```python
MostPopular = data.query('Installs=="50,000,000+"')
MostPopular.shape
```

```python
LeastPopular = data.query('Installs=="5+"')
LeastPopular.shape
```

```python
Labels = 'LeastPopular', 'MostPopular'
sizes = [3, 97]
explode = [0.2,0]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=Labels, autopct='%1.1f%%',
        shadow=False)
ax1.axis('equal')
plt.show()
#This correlation shows that there are a lot more apps on the appstore with a large amount of downloads than very few downloads.
```

```python
plt.figure(figsize=(30,20))
x1 = data.Installs
y1 = data.Price
plt.scatter(x1, y1)
plt.title('Installs vs. Price')
plt.axvline(color='black')
plt.axhline(color='black')
plt.grid()
plt.xlabel("Installs")
plt.ylabel("Price");

plt.show()
#This graph shows that the apps that have extremely large amounts of downloads are all free download, while less popular apps have a lot of variation to their prices.
#This indicates that free apps are far more likely to do better on the app store in terms of downloads.
```
